```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('dataset/Social_Network_Ads.csv')
data
```

```python
data.shape
```

```python
x=data.iloc[:,[2,3]].values
y=data.iloc[:,4].values
```

```python
x
```

```python
y
```

```python
import warnings;
warnings.simplefilter('ignore') 
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
train_x
```

```python
test_x
```

```python
test_x.shape
```

```python
train_x.shape
```

```python
train_y
```

```python
test_y
```

```python
#Perfrom Feature Scaling because input data:- differnce
#between age & salary much more....
# so that solving ths problem optimizing this perform feature scaling...
```

```python
#Load Feture Scaling Library
from sklearn.preprocessing import StandardScaler
```

```python
s_c=StandardScaler()
train_x=s_c.fit_transform(train_x)
test_x=s_c.fit_transform(test_x)
train_x
```

```python
test_x
```

```python
#Load KNN Classifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier()
knn.fit(train_x,train_y)
```

```python
y_pred=knn.predict(test_x)
y_pred
```

```python
test_y
```

```python
#Using Confusion Matrix we can find out in testing data how many
#wrong & Correct prediction data.
# first Load Library of Confusion Matrix
from sklearn.metrics import confusion_matrix
```

```python
c_m=confusion_matrix(test_y,y_pred)
c_m
```

```python
#[64+29=93] is correctly predicted &
#[4+3=7] is worng predicted then accuarcy is 93%
```

# How to Plot a Confusion Matrix:-

```python
sb.heatmap(c_m, annot=True)
plt.show()
```

# Evaluation:-

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(test_y,y_pred)
```

```python
print('Wrong Predictions out of Total:') 
print((test_y!=y_pred).sum(),'/',((test_y==y_pred).sum()+(test_y!=y_pred).sum()))
```

```python
np.where(y_pred!=test_y)
```

# Precision & Recall:-

```python
from sklearn.metrics import precision_score, recall_score
```

# Precision Score

```python
prec_0 = 64/(64+3)
prec_1 = 29/(29+4)
print(prec_0)
print(prec_1)
```

```python
print(precision_score(test_y,y_pred))
print(precision_score(test_y,y_pred,average=None))
```

# Recall Score:-

```python
recall_0 = 64/(64+4)
recall_1= 29/(29+3)
print(recall_0)
print(recall_1)
```

```python
print(recall_score(test_y,y_pred))
```

# F1-Score:
    

```python
from sklearn.metrics import f1_score
```

```python
f1_score(test_y,y_pred)
```

# Classification Report:

```python
from sklearn.metrics import classification_report
print(classification_report(test_y,y_pred))
```
