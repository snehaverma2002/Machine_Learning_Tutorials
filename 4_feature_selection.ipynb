# Feature Selection:-

```python
# How to find out Important feature columns using Feature Slection:-
#4 Ways
#  1. Using Visualization
#  2. Using corr() method [Correlation of Columns]
#  3. Uasing Seaborn Heatmap Plotting
#  4. Using Random Forest Model.
```

```python
#pip install seaborn --upgrade
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris.keys()
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
X=iris.data 
y=iris.target
```

```python
X
```

```python
y
```

```python
plt.scatter(X[:,2],X[:,3],c=y)
plt.show()
#take petal length & petal width columns, because in this columns much less
# missclassified samples.
# then we can pass only 2 importanat columns data in ML feature
#selection rather than passing
# all 4 columns of Iris datasets.
```

```python
plt.scatter(X[:,1],X[:,2],c=y)
plt.show()
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
plt.scatter(X[:,0],X[:,3],c=y)
plt.show()
```

```python
plt.scatter(X[:50,0],X[:50,1],c='red', label='0')
plt.scatter(X[50:100,0],X[50:100,1],c='green', label='1')
plt.scatter(X[100:,0],X[100:,1], c='blue',label='2')
plt.legend()
plt.show()
```

```python
plt.scatter(X[:50,2],X[:50,3],c='red', label='0')
plt.scatter(X[50:100,2],X[50:100,3],c='green', label='1')
plt.scatter(X[100:,2],X[100:,3], c='blue',label='2')
plt.legend()
plt.show()
```

```python
df = pd.DataFrame(X,columns=iris.feature_names)
df['target']=y
df.head()
```

```python
df.corr()
#corr():- Comput pairwise correlation of columns, excluding NA/null values.
#In this case petal length & petal width columns correlation values is
# much higher
#than another column.i.e [0.949035,0.956547], so that take only these columns.
```

```python
 import seaborn as sb
    
```

```python
sb.heatmap(df.corr(),annot=True)
plt.show()
```
